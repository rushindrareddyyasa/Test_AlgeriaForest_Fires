## Algerian Forest Fires Dataset

### Dataset information:
### The dataset includes 244 instances that regroup a data of two regions of algeria,namely the Bejala region located in the northest of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria

### 122 instances of each region

### The period from june 2012 to september 2012.The dataste includes 11 attributs and 1 output attribute have been classified into fire (138 classes) and not fire(106 classes).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

### Dat Cleaning

```python
## Missing values
dataset[dataset.isna().any(axis=1)]
```

### The dataset is converted into two sets based on region from 122th index,we can make a new column based on the region

### 1 : "Bejaia Region Dataset"
### 2 : "Sidi-Bel Abbes Region Dataset"
### Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.iloc[[122]]
```

```python
## Remove the 122nd Row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
### Fix spaces in columns names

df.columns = df.columns.str.strip()
df.columns
```

```python
## Change the requird columns as integer datatype

df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
### Changing other columns to floatdata datatype

objects = [features for features in df.columns if df[features].dtypes == 'O']

for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
## let ave the cleaned dataset

df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
